##Import required Packages

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
# Read the csv file
df = pd.read_csv('/content/Amazon Sales data.csv')
```

```python
df.head(2)
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
```

```python
df.head(2)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df, x =df['Order Year'], y = df['Total Revenue'], color ='orange')
plt.title("Sales Performance Analysis: Trends in units sold 2010-2017")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df, x =df['Order Year'], y = df['Total Profit'], color ='orange')
plt.title("Profit Trends over 2010-2017 analysis")
plt.show()
```

```python
data_preprocess = pd.DataFrame({
    'Year': df['Order Year'],
    'Total Revenue': df['Total Revenue'],
    'Total cost': df['Total Cost']
})
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x ='Year', y ='value', hue='variable',data = pd.melt(data_preprocess,['Year']),errorbar=None)
plt.legend()
plt.grid()
plt.title("Total Revenue vs Total Cost 2010-2017")
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = df, x = df['Order Month'], y = df['Total Profit'])
plt.title("Monthly Total profit: A box Plot Analysis 2010-2017")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x = df['Order Month'], y = df['Total Revenue'])
plt.title("Sales by Month 2010-2017")
plt.show()
```

```python
df_2010 = df[df['Order Year']==2010]
```

```python
df_2010.head()
```

```python
df_2010 = df_2010.sort_values(by='Total Profit', ascending=False)
df_2010.head()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df_2010, x = df_2010['Order Month'], y = df_2010['Total Profit'])
plt.grid()
plt.title("Monthly Profit trends in 2010 A BarPlot")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data = df_2010, x = df_2010['Order Month'], y = df_2010['Total Profit'])
plt.grid()
plt.title("Monthly Profit trends in 2010 A BarPlot")
plt.show()
```

```python
df.sort_values(by = 'Total Profit', ascending=False)
df[df['Country'] == 'Turkmenistan']
```

```python
CountryProfit = df.groupby('Country')['Total Profit'].sum().reset_index()
CountryProfit = CountryProfit.sort_values(by='Total Profit', ascending=False).reset_index()
CountryProfit.head()
```

```python
plt.figure(figsize=(14,14))
sns.set_palette("pastel")
sns.set(style="whitegrid")
plt.title('Total Profit by Country')
sns.barplot(data=CountryProfit, x='Total Profit', y='Country',errorbar=None, hue= 'Total Profit')
plt.show()
```

```python
ProductProfit = df.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False).reset_index()
ProductProfit.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=ProductProfit, x='Total Profit', y='Item Type',errorbar=None, hue= 'Total Profit')
plt.title("Profit by Item Type")
plt.show()
```
